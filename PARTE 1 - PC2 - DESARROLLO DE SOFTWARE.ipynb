```ruby
#FUNCIÓN QUE ACEPTE UNA CADENA QUE CONTENGA TODAS LAS LETRAS DEL ALFABETO
#EXCEPTO UNA Y DEVUELVA LA LETRA QUE FALTA. (CON COMPLEJIDAD O(n))

def encontrar_letra_faltante(cadena)
  alfabeto = ('a'..'z').to_a
  frecuencia = Hash.new(0)

  # Contar la frecuencia de cada letra en la cadena
  cadena.each_char { |letra| frecuencia[letra] += 1 }

  # Buscar la letra que falta
  alfabeto.each do |letra|
    return letra if frecuencia[letra] == 0
  end

  # Si no falta ninguna letra entonces debe devolver nil
  return nil
end

# Hacemos un ejemplo de uso
cadena = "the quick brown box jumps over a lazy dog"
letra_faltante = encontrar_letra_faltante(cadena)
puts "La letra que falta es: #{letra_faltante}"
```

```ruby
##PREGUNTAS 2 Y 3 JUNTAS YA QUE LA IMPLEMENTACIÓN DE LOS MÉTODOS
##EN LA CLASE BYNARYTREE IMPLICA CONSTRUIR UN ÁRBOL BINARIO ORDENADO
##Y LUEGO EXTENDERLA PARA INCLUIR LOS MÉTODOS ADICIONALES

# Clase para representar un nodo en el árbol binario ordenado
class TreeNode
  attr_accessor :value, :left, :right

  def initialize(value)
    @value = value
    @left = nil
    @right = nil
  end
end

# Clase para representar un árbol binario ordenado
class BinaryTree
  include Enumerable

  attr_reader :root

  def initialize
    @root = nil
  end

  # Método para insertar un elemento en el árbol
  def <<(value)
    @root = insert_recursively(@root, value)
  end

  # Método privado para insertar un elemento recursivamente
  def insert_recursively(node, value)
    return TreeNode.new(value) if node.nil?

    if value < node.value
      node.left = insert_recursively(node.left, value)
    else
      node.right = insert_recursively(node.right, value)
    end

    node
  end

  # Método para verificar si el árbol está vacío
  def empty?
    @root.nil?
  end

  # Método each para permitir la iteración sobre los elementos del árbol
  def each(&block)
    traverse_inorder(@root, &block)
  end

  # Método privado para recorrer el árbol en orden (inorder traversal)
  def traverse_inorder(node, &block)
    return if node.nil?

    traverse_inorder(node.left, &block)
    yield node.value
    traverse_inorder(node.right, &block)
  end

  # Método para verificar si el árbol incluye un elemento (extensión)
  def include?(value)
    search_recursively(@root, value)
  end

  # Método privado para buscar un elemento recursivamente
  def search_recursively(node, value)
    return false if node.nil?

    if value == node.value
      return true
    elsif value < node.value
      return search_recursively(node.left, value)
    else
      return search_recursively(node.right, value)
    end
  end

  # Método para verificar si todos los elementos cumplen una condición (extensión)
  def all?(&block)
    all_recursive(@root, &block)
  end

  # Método privado para verificar si todos los elementos cumplen una condición recursivamente
  def all_recursive(node, &block)
    return true if node.nil?

    return false unless yield node.value
    return all_recursive(node.left, &block) && all_recursive(node.right, &block)
  end

  # Método para verificar si algún elemento cumple una condición (extensión)
  def any?(&block)
    any_recursive(@root, &block)
  end

  # Método privado para verificar si algún elemento cumple una condición recursivamente
  def any_recursive(node, &block)
    return false if node.nil?

    return true if yield node.value
    return any_recursive(node.left, &block) || any_recursive(node.right, &block)
  end

  # Método para ordenar los elementos del árbol (extensión)
  def sort
    to_a.sort
  end
end

# Hacemos un ejemplo de uso
tree = BinaryTree.new
tree << 5
tree << 3
tree << 7
tree << 2
tree << 4
tree << 6
tree << 8

# Hacemos un ejemplo de llamadas a métodos extendidos
puts "Include 4: #{tree.include?(4)}"
puts "All elements are even: #{tree.all? { |value| value.even? }}"
puts "Any element is odd: #{tree.any? { |value| value.odd? }}"
puts "Sorted elements: #{tree.sort.to_a}"
```
